```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 50)
```

```python
y = 1.2*x + 0.8
```

```python
plt.scatter(x,y)
plt.plot(x, y, 'r')
```
